```python
import numpy as np
from mvp_fft.mvp_fft import MBT_fft_init
from bl_krylov.bl_krylov import bl_cocg_rq_jacobi_mvp_fft
```

```python
n= np.array([3,3,3],dtype=int)
f=3
lf=1
k=0.1

L=10

itermax= 1000
tol=1e-4
```

```python
## MBT projection of the non-diagonal block of DDA matrix
Au_til= MBT_fft_init(n,f,lf,k)
```

```python
address= np.array([], dtype=int)
for ix in range(n[0]):
    for iy in range(n[1]):
        for iz in range(n[2]):
            lattice_address= n[1]*n[2]*ix+n[2]*iy+iz

            if lattice_address%5==0 : # condition within target volume (hypothetical condition)
                address= np.append(address, lattice_address)


B= np.random.rand(f*len(address),L)
DIAG_A= np.ones(f*len(address))
```

```python
X, iter_fin, err_fin= bl_cocg_rq_jacobi_mvp_fft(n,f,address,Au_til,DIAG_A,B,tol,itermax)
X, iter_fin, err_fin
```
