```python
import pyfftw
import numpy as np
```

```python
a = pyfftw.empty_aligned(128, dtype='complex64')
b = pyfftw.empty_aligned(128, dtype='complex64')

fft_object = pyfftw.FFTW(a, b, direction='FFTW_FORWARD')
```

```python
c = pyfftw.empty_aligned(128, dtype='complex64')
ifft_object = pyfftw.FFTW(b, c, direction='FFTW_BACKWARD')
```

```python
# Generate some data
ar, ai = np.random.randn(2, 128)
a[:] = ar + 1j*ai
```

```python
fft_object.execute()
```
